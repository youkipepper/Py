```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

# 1 腐蚀与膨胀

```python
img = cv.imread("./letter.png")
```

```python
plt.imshow(img[:, :, ::-1])
```

```python
# 创建核结构
kenel = np.ones((5, 5), np.uint8)
```

```python
# 腐蚀
img2 = cv.erode(img, kenel)
```

```python
plt.imshow(img2[:, :, ::-1])
```

```python
# 膨胀
img1 = cv.dilate(img, kenel)
```

```python
plt.imshow(img[:, :, ::-1])
```

# 2 开闭运算

```python
open = cv.imread("./letteropen.png")
```

```python
plt.imshow(open[:, :, ::-1])
```

```python
kenel = np.ones((10, 10), np.uint8)
```

```python
cvOpen = cv.morphologyEx(open, cv.MORPH_OPEN, kenel)
```

```python
plt.imshow(cvOpen[:, :, ::-1])
```

```python
close = cv.imread("./letterclose.png")
```

```python
plt.imshow(close[:, :, ::-1])
```

```python
cvClose = cv.morphologyEx(close, cv.MORPH_CLOSE, kenel)
```

```python
plt.imshow(cvClose[:, :, ::-1])
```

# 3 黑帽和礼帽

```python
top = cv.morphologyEx(open, cv.MORPH_TOPHAT, kenel)
```

```python
plt.imshow(top)
```

```python
black = cv.morphologyEx(close, cv.MORPH_BLACKHAT, kenel)
```

```python
plt.imshow(black)
```
