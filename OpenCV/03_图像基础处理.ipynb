```python
demo = cv.imread("./test.jpg")
```

```python
plt.imshow(demo[:,:,::-1])
```

```python
b, g, r = cv.split(demo)
```

```python
plt.imshow(b, cmap=plt.cm.gray)
```

```python
img2 = cv.merge((b, g, r))
```

```python
plt.imshow(img2[:, :, ::-1])
```

```python
gray = cv.cvtColor(demo, cv.COLOR_BGR2GRAY)
```

```python
plt.imshow(gray, cmap=plt.cm.gray)
```

```python
hsv = cv.cvtColor(demo, cv.COLOR_BGR2HSV)
plt.imshow(hsv)
```
