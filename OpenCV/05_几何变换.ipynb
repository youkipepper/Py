```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
kids = cv.imread("./kids.jpg")
```

```python
plt.imshow(kids[:, :, ::-1])
```

```python
# 绝对尺寸
rows, cols = kids.shape[:2]
```

```python
rows
```

```python
cols
```

```python
res = cv.resize(kids, (2*cols, 2*rows))
```

```python
plt.imshow(res[:, :, ::-1])
```

```python
res.shape
```

```python
res1 = cv.resize(kids, None, fx=0.5, fy=0.5)
```

```python
plt.imshow(res1[:, :, ::-1])
```

```python
res1.shape
```

```python
# 图像平移
plt.imshow(kids[:, :, ::-1])
```

```python
rows
```

```python
cols
```

```python
M = np.float32([[1, 0, 100], [0, 1, 50]])
```

```python
res2 = cv.warpAffine(kids, M, (2*cols, 2*rows))
```

```python
plt.imshow(res2[:, :, ::-1])
```

```python
# 图像旋转
M = cv.getRotationMatrix2D((cols/2, rows/2), 45, 0.5)
```

```python
res3 = cv.warpAffine(kids, M, (cols, rows))
```

```python
plt.imshow(res3[:, :, ::-1])
```

# 4 仿射变换

```python
pts1 = np.float32([[50, 50], [200, 50], [50, 200]])
pts2 = np.float32([[100, 100], [200, 50], [100, 250]])
```

```python
M = cv.getAffineTransform(pts1, pts2)
```

```python
M
```

```python
res4 = cv.warpAffine(kids, M, (cols, rows))
```

```python
plt.imshow(res4[:, :, ::-1])
```
