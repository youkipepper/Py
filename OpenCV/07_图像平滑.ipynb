```python
import numpy as np 
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
dogsp = cv.imread("./dogsp.jpeg")
plt.imshow(dogsp[:, :, ::-1])
```

```python
doggau = cv.imread("./dogGauss.jpeg")
```

# 1 均值滤波 

```python
dog = cv.blur(dogsp, (5, 5))
```

```python
plt.imshow(dog[:, :, ::-1])
```

# 2 高斯滤波

```python
dog = cv.GaussianBlur(doggau, (3, 3), 1)  # 卷积核的取值一般为奇数
```

```python
plt.imshow(dog[:, :, ::-1])
```

```python
dog = cv.medianBlur(dogsp, 3)
```

```python
plt.imshow(dog[:, :, ::-1])
```
