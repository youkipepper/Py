```python
import numpy as np 
import cv2 as cv
import matplotlib.pyplot as plt
```

# 1 霍夫线检测

```python
img = cv.imread("./rili.jpg")
```

```python
edge = cv.Canny(img, 50, 150)
```

```python
plt.imshow(edge, cmap=plt.cm.gray)
```

```python
lines = cv.HoughLines(edge, 0.8, np.pi/180, 150)
```

```python
for line in lines:
    rho, theta = line[0]
    a = np.cos(theta)
    b = np.sin(theta)
    x_0 = rho*a
    y_0 = rho*b
    x_1 = int(x_0+1000*(-b))
    y_1 = int(y_0+1000*a)
    x_2 = int(x_0-1000*(-b))
    y_2 = int(y_0-1000*a)
    cv.line(img, (x_1, y_1), (x_2, y_2), (0, 255, 0))
```
