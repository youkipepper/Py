```python
import matplotlib.pyplot as plt
from matplotlib_inline import backend_inline
backend_inline.set_matplotlib_formats('png')
plt.rcParams[('figure.dpi')] = 300
plt.rcParams[('savefig.dpi')] = 300
```

```python
import numpy as np
import pandas as pd
import scipy.stats as stats
```

```python
mu = 0
sigma = 1

z_scores = np.linspace(-4, 4, 1000)
```

```python
pdf = stats.norm.pdf(z_scores, mu, sigma) 
```

```python
plt.figure(figsize=(9, 5))
plt.plot(z_scores, pdf, label= f'Nomal Distribution\n$\mu$ = {mu}, $\sigma$ = {sigma}')
plt.fill_between(z_scores, pdf, alpha= 0.2)

for z in [-3, -2, -1, 0, 1, 2, 3]:
    plt.axvline(x=z, linestyle= '--', color = 'grey')
    plt.text(z, 0.02, f'Z={z}', horizontalalignment='center')

plt.ylim(0, max(pdf)*1.1)

plt.title('Standard Normal Distribution PDF')
plt.xlabel('Z-Scores')
plt.ylabel('Probability Density')
plt.legend()
plt.grid()
plt.show()
```
